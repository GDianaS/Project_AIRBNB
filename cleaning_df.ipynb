```python
import pandas as pd
```

```python
# Importação dos arquivos csv
df_blm_original = pd.read_csv('database/original/blm_original.csv')
df_stm_original = pd.read_csv('database/original/stm_original.csv')
```

```python
df_blm_highlights = df_blm_original.apply(lambda x: list([x['highlights/0/title'],
                                              x['highlights/1/title'],
                                              x['highlights/2/title'],
                                              x['highlights/3/title']]), axis=1)

df_blm_highlights_host= df_blm_original.apply(lambda x: list([x['host/highlights/0'],
                                              x['host/highlights/1'],
                                              x['host/highlights/2'],
                                              x['host/highlights/3'],
                                              x['host/highlights/4'],
                                              ]), axis=1)

df_blm_highlights_location = df_blm_original.apply(lambda x: list([x['locationDescriptions/0/content'],
                                              x['locationDescriptions/1/content'],
                                              ]), axis=1)

df_blm = df_blm_original.loc[:,['id', 'title','description','subDescription/title', 'roomType', 'price/discountedPrice','price/price', 'price/breakDown/basePrice/price', 'url', 'rating/value','rating/reviewsCount','rating/guestSatisfaction', 'rating/communication', 'rating/cleanliness','rating/accuracy', 'rating/location','host/id','host/name','host/about','host/ratingAverage','host/isSuperHost','host/hostDetails/1','host/hostDetails/0', 'coordinates/latitude','coordinates/longitude']]

df_blm.rename(columns={ 'subDescription/title':'typeDescription',
                        'price/price':'price',
                        'price/discountedPrice':'discountedPrice',
                        'price/breakDown/basePrice/price':'day5Price',
                        'rating/value':'ratingValue',
                        'rating/reviewsCount':'ratingReviewsCount',
                        'rating/guestSatisfaction':'ratingGuestSatisfaction',
                        'rating/communication':'ratingCommunication',
                        'rating/cleanliness':'ratingCleanliness',
                        'rating/accuracy':'ratingAccuracy',
                        'rating/location':'ratingLocation',
                        'host/id':'hostID',
                        'host/name':'hostName',
                        'host/about':'hostAbout',
                        'host/ratingAverage':'hostRating',
                        'host/isSuperHost':'hostIsSuper',
                        'host/hostDetails/1':'hostTimeResponse',
                        'host/hostDetails/0':'hostResponseRate',
                        }, inplace=True)


df_blm = df_blm.assign(highlights=df_blm_highlights)
df_blm = df_blm.assign(highlightsHost = df_blm_highlights_host)
df_blm = df_blm.assign(highlightsLocation = df_blm_highlights_location )
```

```python
df_stm_highlights = df_stm_original.apply(lambda x: list([x['highlights/0/title'],
                                              x['highlights/1/title'],
                                              x['highlights/2/title'],
                                              x['highlights/3/title']]), axis=1)

df_stm_highlights_host= df_stm_original.apply(lambda x: list([x['host/highlights/0'],
                                              x['host/highlights/1'],
                                              x['host/highlights/2'],
                                              x['host/highlights/3'],
                                              x['host/highlights/4'],
                                              ]), axis=1)

df_stm_highlights_location = df_stm_original.apply(lambda x: list([x['locationDescriptions/0/content'],
                                              x['locationDescriptions/1/content'],
                                              ]), axis=1)

df_stm = df_stm_original.loc[:,['id', 'title','description','subDescription/title', 'roomType', 'price/discountedPrice','price/price', 'price/breakDown/basePrice/price', 'url', 'rating/value','rating/reviewsCount','rating/guestSatisfaction', 'rating/communication', 'rating/cleanliness','rating/accuracy', 'rating/location','host/id','host/name','host/about','host/ratingAverage','host/isSuperHost','host/hostDetails/1','host/hostDetails/0', 'coordinates/latitude','coordinates/longitude']]

df_stm.rename(columns={ 'subDescription/title':'typeDescription',
                        'price/price':'price',
                        'price/discountedPrice':'discountedPrice',
                        'price/breakDown/basePrice/price':'day5Price',
                        'rating/value':'ratingValue',
                        'rating/reviewsCount':'ratingReviewsCount',
                        'rating/guestSatisfaction':'ratingGuestSatisfaction',
                        'rating/communication':'ratingCommunication',
                        'rating/cleanliness':'ratingCleanliness',
                        'rating/accuracy':'ratingAccuracy',
                        'rating/location':'ratingLocation',
                        'host/id':'hostID',
                        'host/name':'hostName',
                        'host/about':'hostAbout',
                        'host/ratingAverage':'hostRating',
                        'host/isSuperHost':'hostIsSuper',
                        'host/hostDetails/1':'hostTimeResponse',
                        'host/hostDetails/0':'hostResponseRate',
                        }, inplace=True)


df_stm = df_stm.assign(highlights=df_stm_highlights)
df_stm = df_stm.assign(highlightsHost = df_stm_highlights_host)
df_stm = df_stm.assign(highlightsLocation = df_stm_highlights_location )
```

```python
# Adicionar coluna de origem
df_blm['origin'] = 'blm'
df_stm['origin'] = 'stm'

# Concatenar os dataframes
df_combined = pd.concat([df_blm, df_stm], ignore_index=True)

# Exibir o dataframe resultante
df_combined
```
